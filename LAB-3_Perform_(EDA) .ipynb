# Task
Perform a comprehensive Exploratory Data Analysis (EDA) of the dataset, including an initial overview, missing value analysis, outlier detection, and correlation analysis, and summarize the key findings.

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame for demonstration purposes
data = {
    'feature_A': np.random.rand(100) * 100,
    'feature_B': np.random.randint(1, 10, 100),
    'feature_C': np.random.choice(['X', 'Y', 'Z'], 100),
    'feature_D': np.random.randn(100),
    'target': np.random.randint(0, 2, 100)
}
df = pd.DataFrame(data)

# Introduce some missing values and outliers for demonstration
df.loc[df.sample(frac=0.05).index, 'feature_A'] = np.nan
df.loc[df.sample(frac=0.03).index, 'feature_B'] = np.nan
df.loc[df.sample(frac=0.02).index, 'feature_D'] = 100 # Outlier

print("Sample DataFrame 'df' created successfully.")
```

## Missing Values Analysis

### Subtask:
Identify the presence of missing values in each column of the dataset using `df.isnull().sum()` and visualize the results.


**Reasoning**:
I will calculate the missing values for each column, filter out columns with no missing values, and then visualize the remaining missing values using a bar plot to fulfill the subtask's requirements for missing value analysis.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Calculate the number of missing values for each column
missing_values_count = df.isnull().sum()

# 2. Filter out columns that have zero missing values
missing_values_count = missing_values_count[missing_values_count > 0]

print("Missing values per column:")
print(missing_values_count)

# 3. Create a bar plot to visualize the count of missing values for each column
if not missing_values_count.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=missing_values_count.index, y=missing_values_count.values, palette='viridis')
    plt.xlabel('Columns')
    plt.ylabel('Number of Missing Values')
    plt.title('Missing Values per Column')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("No missing values found in the DataFrame.")
```

## Outlier Detection

### Subtask:
Analyze the distribution of key numerical variables to identify potential outliers using box plots.


**Reasoning**:
To analyze the distribution of key numerical variables and identify potential outliers as requested, I will iterate through the numerical columns of the DataFrame `df` (excluding the 'target' column as it is binary and 'feature_B' since it's ordinal and less prone to traditional outliers in this context). For each selected numerical column, I will generate a box plot to visually represent its distribution and highlight any data points that fall outside the typical range, indicating potential outliers. Finally, I will explicitly show each plot to ensure it is displayed.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numerical columns (excluding 'target' and 'feature_C' which is categorical, and 'feature_B' which is ordinal)
# Based on df.info() and df.describe(), target is binary and feature_C is object type.
# feature_B is integer type with a small range (1-9), making box plot for outlier detection less typical than for continuous data.
# So, focusing on continuous numerical features: feature_A and feature_D.
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

# Exclude binary target and ordinal feature if they exist and are not suitable for continuous outlier analysis
if 'target' in numerical_cols:
    numerical_cols.remove('target')
if 'feature_B' in numerical_cols: # As discussed, feature_B has a small integer range, so it's better to exclude for typical boxplot outlier detection
    numerical_cols.remove('feature_B')

print(f"Analyzing potential outliers for numerical columns: {numerical_cols}")

# Create box plots for each numerical column
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

## Correlation Analysis

### Subtask:
Examine the relationships between numerical variables by calculating and visualizing a correlation matrix using a heatmap.


**Reasoning**:
To analyze the relationships between numerical variables, I will first select the continuous numerical columns 'feature_A' and 'feature_D' from the DataFrame. Then, I will compute their Pearson correlation matrix and visualize it as a heatmap with annotations and a color bar to show the strength and direction of the correlations.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Select only the numerical columns suitable for correlation analysis
# Exclude 'target' (binary) and 'feature_C' (categorical).
# Exclude 'feature_B' as it's ordinal with a small integer range, making Pearson correlation less interpretable.
# The continuous numerical features are 'feature_A' and 'feature_D'.
numerical_cols_for_corr = ['feature_A', 'feature_D']

# Ensure these columns actually exist in the DataFrame
numerical_df = df[numerical_cols_for_corr]

# 2. Calculate the Pearson correlation matrix
correlation_matrix = numerical_df.corr(method='pearson')

print("Correlation Matrix:")
print(correlation_matrix)

# 3. Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, cbar=True)
plt.title('Correlation Matrix of Numerical Features')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Reasoning**:
The previous correlation analysis step only included 'feature_A' and 'feature_D', but the updated instructions explicitly state that 'feature_B' should also be included in the correlation analysis. Therefore, I need to modify the code to include 'feature_B' in the selection of numerical columns before calculating and visualizing the correlation matrix.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Select only the numerical columns suitable for correlation analysis
# Exclude 'target' (binary) and 'feature_C' (categorical).
# As per new instructions, include 'feature_A', 'feature_B', and 'feature_D'.
numerical_cols_for_corr = ['feature_A', 'feature_B', 'feature_D']

# Ensure these columns actually exist in the DataFrame
numerical_df = df[numerical_cols_for_corr]

# 2. Calculate the Pearson correlation matrix
correlation_matrix = numerical_df.corr(method='pearson')

print("Correlation Matrix:")
print(correlation_matrix)

# 3. Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8)) # Adjusted figure size for more columns
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, cbar=True)
plt.title('Correlation Matrix of Numerical Features')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
