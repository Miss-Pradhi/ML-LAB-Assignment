## Gender Distribution Plot

### Subtask:
Create a plot (e.g., bar chart or pie chart) to visualize the proportion of students by gender. Ensure a legend is included.


**Reasoning**:
To visualize the proportion of students by gender, I will use a seaborn countplot, which is a type of bar chart, and then display it using matplotlib.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a count plot for 'gender'
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='gender', palette='viridis')
plt.title('Distribution of Students by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.show()
```

**Reasoning**:
The previous code generated a `FutureWarning` regarding the usage of `palette` without `hue`. To address this, I will explicitly set `hue='gender'` and `legend=False` in the `sns.countplot` function.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a count plot for 'gender'
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='gender', hue='gender', palette='viridis', legend=False)
plt.title('Distribution of Students by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.show()
```

**Reasoning**:
The previous plot successfully addressed the `FutureWarning`, but the subtask explicitly requires a legend. Since `hue='gender'` is already set, removing `legend=False` should automatically generate a legend for the 'gender' categories.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a count plot for 'gender' with a legend
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='gender', hue='gender', palette='viridis')
plt.title('Distribution of Students by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.show()
```

## Scores Relationship Plot

### Subtask:
Generate a scatter plot matrix or pair plot to visualize the relationships between different subject scores (math, reading, writing). Ensure a legend is included.


**Reasoning**:
To visualize the relationships between the subject scores and include a legend differentiating by gender, I will create a pair plot using `seaborn.pairplot()`, as specified in the instructions, setting the `hue` parameter to 'gender' to automatically generate a legend.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a pair plot to visualize relationships between subject scores
# Use 'gender' for hue to include a legend and differentiate points by gender
sns.pairplot(df, vars=['math score', 'reading score', 'writing score'], hue='gender', palette='viridis')

# Add a title to the pair plot
plt.suptitle('Relationships Between Subject Scores by Gender', y=1.02) # Adjust y to prevent title overlap

# Display the plot
plt.show()
```
